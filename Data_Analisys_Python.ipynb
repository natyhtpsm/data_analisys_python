```python
import os
import pandas as pd

total_table = pd.DataFrame()

list_archive = os.listdir("/content/drive/MyDrive/Vendas")
print(list_archive)

for archive in list_archive:
  if "Vendas" in archive:
    table = pd.read_csv(f"/content/drive/MyDrive/Vendas/{archive}")
    total_table = total_table.append(table)

display(total_table)

table_products = total_table.groupby('Produto').sum()
table_products = table_products[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=False)
display(table_products)

total_table['Billing'] = total_table['Quantidade Vendida'] * total_table['Preco Unitario']

table_billing = total_table.groupby('Produto').sum()
table_billing = table_billing[["Billing"]].sort_values(by="Billing", ascending=False)
display(table_billing)

table_stores = total_table.groupby('Loja').sum()
table_stores  = table_stores[['Billing']]
display(table_stores)

import plotly.express as px

graph = px.bar(table_stores, x = table_stores.index, y = 'Billing')
graph.show()

 

```
